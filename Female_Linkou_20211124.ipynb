# Split to training/ valid/ test

```python
X_train = df_3y_F_L[df_3y_F_L.year == 2017].iloc[:, :-2]
y_train = df_3y_F_L[df_3y_F_L.year == 2017].outcome_screen

X_valid = df_3y_F_L[df_3y_F_L.year == 2018].iloc[:, :-2]
y_valid = df_3y_F_L[df_3y_F_L.year == 2018].outcome_screen

X_test = df_3y_F_L[df_3y_F_L.year == 2019].iloc[:, :-2]
y_test = df_3y_F_L[df_3y_F_L.year == 2019].outcome_screen
```

# Model_XGBoost Classifier

```python
start_time = time.time()

xgb = XGBClassifier()
cv = cross_val_score(xgb,X_train,y_train,cv=5)
print(xgboost.__version__)
print(cv)
print('XGBClassifier:', cv.mean())
print("--- %s seconds ---" % (time.time() - start_time))
```

# Training

```python
train = confusio_matrix(xgb, X_train, y_train)
plot_roc_curve(xgb, X_train, y_train)

train = train.rename(index={0: "F_L_HPA+CGRD_train"})
print('--------------- True/False Pos/Neg ---------------')
print(train, '\n')

train_imp = get_impotances(xgb, X_train, y_train)
train_imp.columns = ['F_L_HPA+CGRD_column', 'F_L_train_impor']

# train_imp
```

# Validation

```python
valid = confusio_matrix(xgb, X_valid, y_valid)
plot_roc_curve(xgb, X_valid, y_valid)

valid = valid.rename(index={0: "F_L_HPA+CGRD_valid"})
print('--------------- True/False Pos/Neg ---------------')
print(valid, '\n')

valid_imp = get_impotances(xgb, X_valid, y_valid)
valid_imp.columns = ['F_L_HPA+CGRD_column', 'F_L_valid_impor']
# valid_imp
```

# Test

```python
test = confusio_matrix(xgb, X_test, y_test)
plot_roc_curve(xgb, X_test, y_test)

test = test.rename(index={0: "F_L_HPA+CGRD_test"})
print('--------------- True/False Pos/Neg ---------------')
print(test, '\n')

test_imp = get_impotances(xgb, X_test, y_test)
test_imp.columns = ['F_L_HPA+CGRD_column', 'F_L_test_impor']
# test_imp
```

# Result

```python
print('--- Female_Linkou HPA+CGRD result---')
result = pd.concat([train, valid, test])
result 
```
